SERIES 

Matriz etiquetada *unidimensional* (SIEMPRE) capaz de contener cualquier tipo de datos.

s = pd.Series(data,index = index)

--> Si no pongo indices, el los pone por defecto (inicia desde 0 y termina en n-1)
--> Data puede ser: un diccionario de python, un ndarray, un valor escalar
--> Si data es un ndarray, el índice debe tener la misma longitud que los datos
--> Los indices pueden ser: numeros random, un nombre, una variable

```python
s = pd.Series(np.random.randn(5), index =["a","b","c","d","e"])
s
```
